# Visualization 

Create a bar chart to show the distribution of passengers by class(1st, 2nd, 3rd)

```python
class_distribution = titanic['Pclass'].value_counts().sort_index()
```

```python
class_distribution.plot(kind='bar',color='orange',edgecolor='black')
#Set Title
plt.title('The distribution of passengers by Class')
#Set x-axis label
plt.xlabel('Class')
#Set y-axis label
plt.ylabel('Number of Passengers')
plt.xticks(rotation=360, horizontalalignment='right')
#Display the plot
plt.show()
```

Calculate and visualise the survival rate of passengers. Create a pie chart to
represent this information

```python
survival_counts = titanic['Survived'].value_counts()
total_passengers = len(titanic)
survival_rate = (survival_counts[1] / total_passengers) * 100
```

```python
plt.figure(figsize=(6, 6))
labels = ['Survived','Did Not Survive']
sizes = [survival_counts[1], survival_counts[0]]
colors = ['yellow','green']
explode = (0.1,0)
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, explode=explode)
plt.title(f'Survival Rate')
plt.axis('equal')
plt.show()
```

Create a bar chart to show the count of passengers by gender

```python
gender_distribution = titanic['Sex'].value_counts().sort_index()
```

```python
gender_distribution.plot(kind='bar',color='blue',edgecolor='black')
#Set Title
plt.title('The count of passenger by Gender')
#Set x-axis label
plt.xlabel('Gender')
#Set y-axis label
plt.ylabel('Number of Passengers')
plt.xticks(rotation=360, horizontalalignment='right')
#Display the plot
plt.show()
```

Create a stacked bar chart to visualise the survival rate by gender. Explain any
gender-based differences in survival.

```python
gender_survival_counts = titanic.groupby(['Sex', 'Survived']).size().unstack().fillna(0)
gender_survival_counts.plot(kind='bar', stacked=True, color=['red', 'green'], edgecolor='black')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.legend(['Did Not Survive', 'Survived'], loc='upper right')
plt.show()
```

Explanation:
In the chart, it is revealed that the red portion of the bar indicates passengers who did not survive while green portion indicated Survived.
It is also observed that the survival rate for females appear to be much higher compared to males. Thus indicating noticeable gender-based difference in survival.
